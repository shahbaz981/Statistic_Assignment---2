```python
#Q1. What are the three measures of central tendency?

'''[Q1-answer:-]
(1) Mean: The mean is the average value of a set of numbers. It is calculated by adding up all the values in the dataset and
then dividing by the total number of values.

(2) Median: The median is the middle value in a dataset when the values are arranged in ascending or descending order.
If there is an even number of values, the median is the average of the two middle values.

(3) Mode: The mode is the value that appears most frequently in a dataset. It is possible for a dataset to have one mode,
more than one mode (multimodal), or no mode at all (no value appears more than once).
'''

```

```python
#Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?

'''[Q2-answer:-]
The mean, median, and mode are all measures of central tendency, which means they aim to represent the typical or 
central value of a dataset. Here's how they differ and how they are used:

(1) Mean: The mean is the sum of all the values in a dataset divided by the total number of values. It's essentially the
average value. It's sensitive to extreme values (outliers) in the dataset because it takes into account every value.
The formula for the mean is:

                            Mean = (Sum of all values)/(Number of values)

(2) Median: The median is the middle value when the data set is arranged in ascending or descending order. If there is an
even number of values, the median is the average of the two middle values. The median is not affected by extreme values or
outliers. It's useful when the data is skewed or when extreme values are present. To calculate the median,
you sort the data and find the middle value(s).

                            Median = (Middle value(s))/2

Mode: The mode is the value that appears most frequently in the dataset. Unlike the mean and median, the mode can be applied
to categorical data as well. A dataset can have one mode (unimodal), more than one mode (multimodal), or no mode
at all (if all values occur with the same frequency). The mode is useful for finding the most common value(s) in a dataset.

                            Mode = Most frequent value(s)

'''
```

```python
#Q3. Measure the three measures of central tendency for the given height data:


'''[Q3-answer:-]'''
import numpy as np
from scipy import stats as st

age_list = [178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]

#mean
print('Mean of age_list =',np.mean(age_list))

#Median
print('Median of age_list =',np.median(age_list))

#Mode
frequency = {}
for i in age_list:
    frequency.setdefault(i, 0)
    frequency[i]+=1

frequent = max(frequency.values())
for i, j in frequency.items():
    if j == frequent:
        mode = i
print('Mode of age_list =',mode)
```

```python
#Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset? Provide an example.

'''[Q5-answer:-]
Measures of dispersion, such as range, variance, and standard deviation, are used to describe the spread or variability of 
data points within a dataset. They provide insight into how much the individual data points differ from the central tendency measures like the mean or median.

Here's how each of these measures is used:

(1) Range: The range is the simplest measure of dispersion and represents the difference between the maximum and minimum
values in the dataset. It gives an idea of how spread out the data points are. However, it doesn't consider the distribution
of values between the minimum and maximum. A larger range indicates greater variability in the dataset.

(2) Variance: The variance measures the average squared deviation of each data point from the mean of the dataset.
It gives a more comprehensive understanding of the variability by considering the differences between each data point and
the mean. A higher variance indicates more spread-out data points, while a lower variance suggests that the data points 
are closer to the mean.

(3) Standard Deviation: The standard deviation is the square root of the variance. It provides a measure of the average
distance between each data point and the mean. Like the variance, a higher standard deviation indicates greater variability,
while a lower standard deviation suggests less variability. The standard deviation is often preferred over the variance
because it is in the same units as the original data.

Here's an example illustrating the use of these measures:

Consider the following dataset representing the daily temperatures (in degrees Celsius) recorded over a week: {20,22,18,25,23,21,19}

(1) Range: The range is the difference between the maximum and minimum values:
                Range = 25 − 18 = 7

(2) Variance: To calculate the variance, first, find the mean:
                Mean = (20+22+18+25+23+21+19)
                Mean = 148/7 ≈ 21.14

    Then, calculate the squared differences from the mean and find their average:
    
                Variance = ((20−21.14)^2 + (22−21.14)^2 +...+ (19−21.14)^2)/7
                Variance ≈ 5.33


Standard Deviation: The standard deviation is the square root of the variance:
Standard Deviation = (5.33)^(1/2) ≈ 2.31

So, for this dataset:

Range = 7
Variance ≈ 5.33
Standard Deviation ≈ 2.31
'''
import numpy as np
import seaborn as sns
df = sns.load_dataset('tips')
print('------------variance------------ \n\n',np.var(df))
print('\n--------standard deviation----------- \n\n',np.std(df))
```

```python
#Q6. What is a Venn diagram?

'''[Q6-answer:-]
A Venn diagram is a graphical representation used to show the relationships between different sets or groups of items.
It consists of overlapping circles or other shapes, each representing a set, with the overlapping regions indicating the
intersections between the sets.

Key features of a Venn diagram:

    (1) Sets: Each circle or shape in a Venn diagram represents a set of items or elements.

    (2) Overlapping Regions: The overlapping parts of the circles represent the elements that are common to two or more sets.

    (3) Non-overlapping Regions: The parts of the circles that do not overlap represent elements that belong exclusively to a particular set.

Venn diagrams are commonly used in mathematics, logic, statistics, and various other fields to illustrate concepts such as set
theory, logical relationships, and data classification. They provide a visual and intuitive way to understand the
relationships and intersections between different sets of data or concepts.

For example, consider three sets: A, B, and C. A Venn diagram representing the relationships between these sets
might have three circles overlapping in various ways to show the elements that belong to each set individually,
as well as the elements that are common to two or more sets.
'''
```

```python
#Q8. What do you understand about skewness in data?

'''[Q8-answer:-]
Skewness in data refers to the asymmetry or lack of symmetry in the distribution of values within a dataset.
It indicates the degree to which the data deviates from a symmetrical distribution around the mean.

There are three main types of skewness:

    (1) Positive Skewness (Right Skewness): In a positively skewed distribution, the tail of the distribution extends
        to the right, and the majority of the data points are concentrated on the left side of the distribution. 
        This means that there are more extreme values on the right side of the distribution, which pulls the mean to
        the right of the median. Positive skewness often occurs when there are outliers with relatively large values.

    (2) Negative Skewness (Left Skewness): In a negatively skewed distribution, the tail of the distribution extends
        to the left, and the majority of the data points are concentrated on the right side of the distribution. 
        This means that there are more extreme values on the left side of the distribution, which pulls the mean to
        the left of the median. Negative skewness often occurs when there are outliers with relatively small values.

    (3) Zero Skewness: A distribution is considered to have zero skewness when it is perfectly symmetrical, meaning that
        the left and right sides of the distribution are mirror images of each other. In this case, the mean, median, and
        mode are all equal.
'''
```

```python
#Q12. For a normal distribution data what is the relationship between its measure of central tendency?

'''[Q12-answer:-]
For a normal distribution, the relationship between its measures of central tendency (mean, median, and mode) is as follows:

(1) Mean: In a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.
    This is a characteristic property of normal distributions, and it holds true regardless of the shape, spread, or scale 
    of the distribution. Therefore, for a normal distribution, the mean is equal to the median, which is also equal to the
    mode.

(2) Median: As mentioned, the median of a normal distribution is equal to the mean and mode. This is because the normal 
    distribution is symmetric around its mean, so the median, which represents the midpoint of the distribution, coincides
    with the mean.

(3) Mode: Similarly, the mode of a normal distribution is also equal to the mean and median. In a normal distribution,
    there is only one peak, and it occurs at the mean. Therefore, the value with the highest frequency (mode) is the same as
    the mean and median.

In summary, for a normal distribution:

                                                        Mean = Median = Mode
'''
```

```python
#Q13. How is covariance different from correlation?

'''[Q13-answer:-]
Covariance and correlation are both measures used to quantify the relationship between two variables in a dataset,
but they differ in several key aspects:

(1) Definition:

            * Covariance: Covariance measures the degree to which two variables change together. It indicates the
                direction of the linear relationship between the variables (positive or negative) but does not provide 
                information about the strength or magnitude of the relationship.
            * Correlation: Correlation measures the strength and direction of the linear relationship between two variables. 
                It is a standardized measure that always falls between -1 and 1, where a correlation of 1 indicates a perfect
                positive linear relationship, -1 indicates a perfect negative linear relationship, and 0 indicates no linear
                relationship.
                
(2) Scale:

            * Covariance: Covariance can take any value, positive, negative, or zero, and its units are the product of the
                units of the two variables being measured.
            * Correlation: Correlation is a dimensionless quantity, meaning it does not have any units. It is a standardized
                measure and always falls between -1 and 1.
                
(3) Interpretation:

            * Covariance: Covariance can be difficult to interpret because its value depends on the scale of the variables
                being measured. A positive covariance indicates that the variables tend to move together, a negative
                covariance indicates that the variables move in opposite directions, and a covariance of zero indicates no
                linear relationship.
            * Correlation: Correlation is easier to interpret because it is standardized and always falls between -1 and 1.
                A correlation of 1 or -1 indicates a perfect linear relationship, while a correlation close to 0 indicates a
                weak or no linear relationship.
                
(4) Use:

            * Covariance: Covariance is used to understand how two variables change together but does not provide a 
                standardized measure of the strength of the relationship.
            *Correlation: Correlation is widely used in statistical analysis because it provides a standardized measure of the strength and direction of the linear relationship between two variables. It is particularly useful for identifying associations between variables, assessing the reliability of predictions, and identifying potential issues such as multicollinearity in regression analysis.
'''
```
